```python
def scrape_data(urls):
    data = []

    for url in urls:
        try:
            response = requests.get(url)
            html = response.text
            soup = BeautifulSoup(html, 'html.parser')

            zipcode = soup.select('#header-box > h1.title')[0].text
            median_age = soup.select('a.stat.number span.value')[0].get_text(strip=True).split('±')[0]
            per_capita_income = int(soup.select('.stat.dollar > .primary > .value')[0].text.split('\n')[2].replace(',', '').replace(' ', '').replace('$', ''))
            median_household_income = int(soup.select('.stat.dollar > .primary > .value')[1].text.split('\n')[2].replace(',', '').replace(' ', '').replace('$', ''))

            data.append({
                'Zipcode': zipcode,
                'Median Age': median_age,
                'Per Capita Income': per_capita_income,
                'Median Household Income': median_household_income
            })
        except:
            print(f"Error occurred while scraping data from {url}. Skipping to the next URL.")

    merged_data = pd.DataFrame(data)
    return merged_data
```

```python
urls = ['https://censusreporter.org/profiles/86000US11215-11215/',
        'https://censusreporter.org/profiles/86000US11217-11217/',
        'https://censusreporter.org/profiles/86000US11218-11218/',
        'https://censusreporter.org/profiles/86000US11225-11225/',
        'https://censusreporter.org/profiles/86000US11226-11226/',
        'https://censusreporter.org/profiles/86000US11238-11238/']
```

```python
scraped_data = scrape_data(urls)
```
