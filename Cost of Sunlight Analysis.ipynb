```python
ids = []
zipcode = []
unit_number = []
beds = []
size = []
rents = []
light_cat = []
lights = []

for dictionary in dictionaries:
    for id in dictionary['data']['searchResult']['poi']:
        for insight in id['insights']['insights']:
            if insight['category'] == 'naturalLight':
                ids.append(id['id'])
                zipcode.append(id['addressDetails']['zipcode'])
                unit_number.append(id['addressDetails']['unitNumber'])
                beds.append(id['beds'])
                size.append(id['area'])
                rents.append(id['price'])
                light_cat.append(insight['tradeoff']['value'])
                lights.append(insight['tradeoff']['value'] + f" and the rent is {id['price']}")
        if len(id['insights']['insights']) == 0 or all(insight['category'] != 'naturalLight' for insight in id['insights']['insights']):
            ids.append(id['id'])
            zipcode.append(id['addressDetails']['zipcode'])
            unit_number.append(id['addressDetails']['unitNumber'])
            beds.append(id['beds'])
            size.append(id['area'])
            rents.append(id['price'])
            light_cat.append(f"No sunlight")
            lights.append(f"No sunlight data available and the rent is {id['price']}")
```

## Pre-processing apartment listings with sun exposure details

```python
data = {
    'listing_path': ids,
    'location': zipcode,
    'unit_number': unit_number,
    'beds': beds,
    'size':size,
    'rent': rents,
    'sun_exposure_cat': light_cat,
    'sun_exposure': lights
}

df_sun_exposure = pd.DataFrame(data)
```

```python
df_acs_selected_zipcodes = df_acs_selected_zipcodes.copy()
df_acs_selected_zipcodes.loc[:, 'total'] = pd.to_numeric(df_acs_selected_zipcodes['total'], errors='coerce')

# Calculate the percentage of each living arrangement category
living_arrangements_cols = ['lives alone.1', 'householder living with spouse or spouse of householder.1',
                            'householder living with unmarried partner or unmarried partner of householder.1',
                            'child of householder.1', 'other relatives.1', 'other nonrelatives.1']
df_acs_selected_zipcodes_percent = df_acs_selected_zipcodes.copy()
df_acs_selected_zipcodes_percent.loc[:, living_arrangements_cols] = (
    df_acs_selected_zipcodes_percent[living_arrangements_cols].div(df_acs_selected_zipcodes_percent['18 to 34 years:'], axis=0) * 100
).astype(int)
```

```python
melted = pd.melt(df_acs_selected_zipcodes_percent, id_vars='zipcode', value_vars=living_arrangements_cols, var_name='living_arrangement', value_name='percentage')

# Create the stacked bar chart
chart = alt.Chart(melted).mark_bar().encode(
    x=alt.X('zipcode:N', axis=alt.Axis(title='Zip Code')),
    y=alt.Y('percentage:Q', axis=alt.Axis(title='Percentage')),
    color=alt.Color('living_arrangement:N', legend=alt.Legend(title='Living Arrangement')),
    tooltip=['zipcode', 'living_arrangement', 'percentage']
).properties(
    width=800,
    height=400
)

# Display the chart
chart
```
